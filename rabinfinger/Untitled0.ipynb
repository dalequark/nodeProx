```python
import os


file1 = "./video1"
file2 = "./video2"

os.system("./a.out " + file1 + " > out1")
os.system("./a.out " + file2 +" > out2")

file1 = open("./out1")
file2 = open("./out2")


def getFingerArray(filek):
    i = 0;
    fingerArray = {};
    lastFinger = None;
    content = "";
    for line in filek:
        ar = line.split(',&')
        try:
            if(ar[0][0:len("start:")] == "start:"):
                if lastFinger:
                    fingerArray[lastFinger] = content;
                    content = ar[3]
                lastFinger = ar[1]
            else:
                for data in ar:
                    content += data.rstrip('\n')
        except:
            for data in ar:
                content += data.rstrip('\n')
    return fingerArray
                
a = getFingerArray(file1);
b = getFingerArray(file2);

commonFingers = 0;
for finger in a:
    if finger in b:
        print "a: %s \n b: %s\n\n" % (a[finger], b[finger])
        commonFingers += 1

print "%d common fingers of %d" % (commonFingers, len(a) + len(b) - commonFingers)
```
